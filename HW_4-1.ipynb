## Q1. (10 points) Bayes Optimal Classifier

Consider a one-dimensional dataset  $(X)$ that consists of data from three different classes - $\{c_1, c_2, c_3\}$.  Data from class $c_1$ are distributed according to $Uniform[-1,1]$, those from $c_2$ are distributed according to $Uniform[-2,2]$ and those from $c_3$ are distributed according to $Uniform[-4,4]$.
The priors for the three classes are:
$\mathbb{P} (c_1) = \frac{1}{2}$,   $\mathbb{P} (c_2) = \frac{1}{3}$,
  $\mathbb{P} (c_3) = \frac{1}{6}$.

Now, we want to build a 3-class classifier based on bayes optimal decision theory to predict the most likely label for any given data point.

a. **(2 points)** Write down the class-conditional data distributions for each of the classes defined above.

b. **(4 points)**. Derive the Bayes optimal decision boundaries for this problem.

c. **(4 points)** What is the Bayes Error rate for this classifier?

For each class, we have the following class-conditional data distributions based on the given information:

1. Class 𝑐1:
   Data is distributed according to a Uniform distribution within the range [-1, 1].
   The probability density function (PDF) for class 𝑐1 is:
   \[f_{\text{𝑐1}}(x) = \begin{cases} \frac{1}{2} & \text{if } -1 \leq x \leq 1 \\ 0 & \text{otherwise} \end{cases}\]

2. Class 𝑐2:
   Data is distributed according to a Uniform distribution within the range [-2, 2].
   The probability density function (PDF) for class 𝑐2 is:
   \[f_{\text{𝑐2}}(x) = \begin{cases} \frac{1}{4} & \text{if } -2 \leq x \leq 2 \\ 0 & \text{otherwise} \end{cases}\]

3. Class 𝑐3:
   Data is distributed according to a Uniform distribution within the range [-4, 4].
   The probability density function (PDF) for class 𝑐3 is:
   \[f_{\text{𝑐3}}(x) = \begin{cases} \frac{1}{8} & \text{if } -4 \leq x \leq 4 \\ 0 & \text{otherwise} \end{cases}\]

These class-conditional distributions represent the likelihood of observing a data point given its class.

The next steps in building the Bayes optimal classifier involve calculating the posterior probabilities and making predictions based on the priors and these class-conditional distributions. If you have further questions or need assistance with those steps, please feel free to ask.

To derive the Bayes optimal decision boundaries for this 3-class classification problem, we need to determine the regions in the feature space where each class is most likely. This is done by finding the boundaries at which the posterior probabilities are equal, i.e., the decision boundaries where the probability of belonging to two different classes is the same. We will consider all pairwise combinations of classes.

Let's derive the decision boundaries for the three class combinations:

1. Decision boundary between Class 𝑐1 and Class 𝑐2:
   We want to find the boundary where the posterior probabilities for 𝑐1 and 𝑐2 are equal. Using Bayes' theorem, the posterior probabilities for each class can be calculated as follows:

   For Class 𝑐1:
   \[P(𝑐1 | x) = \frac{P(x | 𝑐1) * P(𝑐1)}{P(x)}\]
   \[P(𝑐1 | x) = \frac{f_{\text{𝑐1}}(x) * P(𝑐1)}{P(x)}\]

   For Class 𝑐2:
   \[P(𝑐2 | x) = \frac{f_{\text{𝑐2}}(x) * P(𝑐2)}{P(x)}\]

   Setting \(P(𝑐1 | x) = P(𝑐2 | x)\) and solving for x:

   \[\frac{f_{\text{𝑐1}}(x) * P(𝑐1)}{P(x)} = \frac{f_{\text{𝑐2}}(x) * P(𝑐2)}{P(x)}\]

   We can ignore \(P(x)\) as it's a common factor on both sides, and we are looking for the boundary, so the specific value of \(P(x)\) doesn't matter. This simplifies to:

   \[f_{\text{𝑐1}}(x) * P(𝑐1) = f_{\text{𝑐2}}(x) * P(𝑐2)\]

   Plugging in the distributions and priors from the problem statement:

   \[\frac{1}{2} * \frac{1}{2} = \frac{1}{4} * \frac{1}{3}\]

   Solve for x:

   \[\frac{1}{4} = \frac{1}{12}\]
   \[x = -1\]

   So, the decision boundary between Class 𝑐1 and Class 𝑐2 is at \(x = -1\).

2. Decision boundary between Class 𝑐1 and Class 𝑐3:
   Following the same procedure as above, we have:

   \[f_{\text{𝑐1}}(x) * P(𝑐1) = f_{\text{𝑐3}}(x) * P(𝑐3)\]

   Plugging in the distributions and priors:

   \[\frac{1}{2} * \frac{1}{2} = \frac{1}{8} * \frac{1}{6}\]

   Solve for x:

   \[\frac{1}{4} = \frac{1}{48}\]
   \[x = -3\]

   So, the decision boundary between Class 𝑐1 and Class 𝑐3 is at \(x = -3\).

3. Decision boundary between Class 𝑐2 and Class 𝑐3:
   Similarly:

   \[f_{\text{𝑐2}}(x) * P(𝑐2) = f_{\text{𝑐3}}(x) * P(𝑐3)\]

   Plugging in the distributions and priors:

   \[\frac{1}{4} * \frac{1}{3} = \frac{1}{8} * \frac{1}{6}\]

   Solve for x:

   \[\frac{1}{12} = \frac{1}{48}\]
   \[x = 1\]

   So, the decision boundary between Class 𝑐2 and Class 𝑐3 is at \(x = 1\).

These decision boundaries divide the feature space into regions where one class is more likely than the other. Data points falling on either side of these boundaries will be classified into the corresponding class.

The Bayes error rate for a classifier represents the lowest possible error rate that can be achieved, and it's the error rate that remains when using the optimal decision boundaries. In this case, we have derived the Bayes optimal decision boundaries for the 3-class classifier. To calculate the Bayes error rate, we need to find the probability of misclassification for each class and then calculate the weighted average of these probabilities based on the class priors.

Let's calculate the Bayes error rate for each class:

1. Class 𝑐1:
   - The probability of misclassification for Class 𝑐1 is the probability that a data point from Class 𝑐1 is incorrectly classified as either Class 𝑐2 or Class 𝑐3.
   - Misclassification can occur if \(x < -1\) (misclassified as Class 𝑐2) or \(x < -3\) (misclassified as Class 𝑐3).
   - Using the class-conditional distributions, we can calculate these probabilities:
     \[P(\text{Misclassification} | \text{𝑐1}) = \int_{-\infty}^{-1} f_{\text{𝑐1}}(x) dx + \int_{-\infty}^{-3} f_{\text{𝑐1}}(x) dx\]
   - Plugging in the distribution for Class 𝑐1:
     \[P(\text{Misclassification} | \text{𝑐1}) = \frac{1}{2} + 0\] (since \(f_{\text{𝑐1}}(x) = 0\) for \(x < -1\))
   - So, \(P(\text{Misclassification} | \text{𝑐1}) = \frac{1}{2}.

2. Class 𝑐2:
   - The probability of misclassification for Class 𝑐2 is the probability that a data point from Class 𝑐2 is incorrectly classified as either Class 𝑐1 or Class 𝑐3.
   - Misclassification can occur if \(x \geq -1\) and \(x < 1\) (misclassified as Class 𝑐1) or \(x < -3\) (misclassified as Class 𝑐3).
   - Using the class-conditional distributions:
     \[P(\text{Misclassification} | \text{𝑐2}) = \int_{-1}^{1} f_{\text{𝑐2}}(x) dx + \int_{-\infty}^{-3} f_{\text{𝑐2}}(x) dx\]
   - Plugging in the distribution for Class 𝑐2:
     \[P(\text{Misclassification} | \text{𝑐2}) = \frac{1}{2} + 0\] (since \(f_{\text{𝑐2}}(x) = 0\) for \(x < -1\))
   - So, \(P(\text{Misclassification} | \text{𝑐2}) = \frac{1}{2}.

3. Class 𝑐3:
   - The probability of misclassification for Class 𝑐3 is the probability that a data point from Class 𝑐3 is incorrectly classified as either Class 𝑐1 or Class 𝑐2.
   - Misclassification can occur if \(x \geq -1\) and \(x < 1\) (misclassified as Class 𝑐1) or \(x \geq 1\) (misclassified as Class 𝑐2).
   - Using the class-conditional distributions:
     \[P(\text{Misclassification} | \text{𝑐3}) = \int_{-1}^{1} f_{\text{𝑐3}}(x) dx + \int_{1}^{\infty} f_{\text{𝑐3}}(x) dx\]
   - Plugging in the distribution for Class 𝑐3:
     \[P(\text{Misclassification} | \text{𝑐3}) = \frac{1}{4} + 0\] (since \(f_{\text{𝑐3}}(x) = 0\) for \(x < 1\))
   - So, \(P(\text{Misclassification} | \text{𝑐3}) = \frac{1}{4}.

Now, to calculate the overall Bayes error rate, we need to take into account the class priors:

- \(ℙ(𝑐1) = \frac{1}{2}\)
- \(ℙ(𝑐2) = \frac{1}{3}\)
- \(ℙ(𝑐3) = \frac{1}{6}\)

The Bayes error rate (\(P(\text{Error})\)) is the weighted sum of the class-specific error probabilities:

\[P(\text{Error}) = ℙ(𝑐1) \cdot P(\text{Misclassification} | \text{𝑐1}) + ℙ(𝑐2) \cdot P(\text{Misclassification} | \text{𝑐2}) + ℙ(𝑐3) \cdot P(\text{Misclassification} | \text{𝑐3})\]

Plugging in the calculated values:

\[P(\text{Error}) = \frac{1}{2} \cdot \frac{1}{2} + \frac{1}{3} \cdot \frac{1}{2} + \frac{1}{6} \cdot \frac{1}{4}\]

Simplify:

\[P(\text{Error}) = \frac{1}{4} + \frac{1}{6} + \frac{1}{24} = \frac{9}{24} = \frac{3}{8}\]

So, the Bayes error rate for this classifier is \(P(\text{Error}) = \frac{3}{8}\).

c. **(5 points)** Build a multiple layer perceptron with the following sequential structure:


*   Fully connected layer mapping from 32 * 32 * 3 dimensions to 50 dimensions
*   ReLU activation
*   Fully connected layer mapping from 50 dimensions to 50 dimensions
*   ReLU activation
*   Fully connected layer mapping from 50 dimensions to 10 dimensions

Then, create an model with the MLP class and an Adam optimizer that will be optimizing the MLP's parameters. Set the initial learning rate to be 0.001.

Tutorial reference:
*   [Build the neural network](https://pytorch.org/tutorials/beginner/basics/buildmodel_tutorial.html)




```python
class MLP(nn.Module):
    def __init__(self):
        super(MLP, self).__init__()
        # Create fully connected layers (nn.Linear)
        ### START CODE ###
        self.fc1 = nn.Linear(32 * 32 * 3, 50)
        self.relu1 = nn.ReLU()
        self.fc2 = nn.Linear(50, 50)
        self.relu2 = nn.ReLU()
        self.fc3 = nn.Linear(50, 10)
        ### END CODE ###

    def forward(self, x):
        # Flatten the input x and pass it through the neural network
        x = x.view(-1, 32*32*3)
        ### START CODE ###
        x = x.view(-1, 32 * 32 * 3)
        x = self.fc1(x)
        x = self.relu1(x)
        x = self.fc2(x)
        x = self.relu2(x)
        x = self.fc3(x)
        ### END CODE ###
        return x
```

```python
import torch.nn as nn
import torch.optim as optim
model = MLP()
# Create the optimizer
### START CODE ###
optimizer = optim.Adam(model.parameters(), lr=0.001)
### END CODE ###
```

d. **(5 points)** Write training and evaluation subroutines per epoch (one full pass of the dataset). Then, train the model for 20 epochs using these subroutines.

Tutorial reference:
*   [Optimizing model parameters](https://pytorch.org/tutorials/beginner/basics/optimization_tutorial.html)



```python
def train(model, loader, optimizer):
    model.train()
    total_loss = 0
    total_num = 0
    for data, target in tqdm(loader):
        out = model(data)
        # Calculate loss based on model output and target
        ### START CODE ###
        loss = nn.CrossEntropyLoss()(out, target)
        ### END CODE ###

        # Use the optimizer to perform backpropagation
        ### START CODE ###
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        ### END CODE ###

        batch_size = len(target)
        total_loss += loss.item() * batch_size
        total_num += batch_size
    avg_loss = total_loss / total_num
    return avg_loss

@torch.no_grad()
def eval(model, loader):
    model.eval()
    total_loss = 0
    total_correct = 0
    total_num = 0
    for data, target in tqdm(loader):
        out = model(data)
        # Calculate loss based on model output and target
        ### START CODE ###
        loss = nn.CrossEntropyLoss()(out, target)
        ### END CODE ###

        # Get model's prediction
        ### START CODE ###
        pred = out.argmax(dim=1)
        ### END CODE ###

        # Count number of correct predictions
        ### START CODE ###
        correct = (pred == target).sum().item()
        ### END CODE ###

        total_correct += correct
        batch_size = len(target)
        total_loss += loss.item() * batch_size
        total_num += batch_size
    avg_loss = total_loss / total_num
    acc = total_correct / total_num
    return avg_loss, acc
```

```python
epochs = 15
for e in range(1, epochs + 1):
    train_loss = train(model, train_loader, optimizer)
    val_loss, val_acc = eval(model, val_loader)
    print(f"Epoch: {e} Train Loss: {train_loss} Val Loss: {val_loss} Val Acc: {val_acc}")
```

Use both the validation and the test data to plot these curves as well as to calculate the AUC-ROC scores. Use one graph for validation data and one graph for test data (all models for validation data in one graph and all models for test data in one graph - and use this scheme for plotting in all subsequent plotting questions).

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score

# Create a single plot for the ROC curves of all three classifiers on the validation set
fig, ax = plt.subplots(figsize=(8, 6))

# Decision Tree
y_pred_decision_tree = decision_tree.predict_proba(X_val)[:, 1]
fpr_decision_tree, tpr_decision_tree, _ = roc_curve(y_val, y_pred_decision_tree)
auc_decision_tree = roc_auc_score(y_val, y_pred_decision_tree)
ax.plot(fpr_decision_tree, tpr_decision_tree, lw=2, label=f'Decision Tree (AUC = {auc_decision_tree:.2f})')

# Logistic Regression
y_pred_logistic_regression = logistic_regression.predict_proba(X_val)[:, 1]
fpr_logistic_regression, tpr_logistic_regression, _ = roc_curve(y_val, y_pred_logistic_regression)
auc_logistic_regression = roc_auc_score(y_val, y_pred_logistic_regression)
ax.plot(fpr_logistic_regression, tpr_logistic_regression, lw=2, label=f'Logistic Regression (AUC = {auc_logistic_regression:.2f})')

# MLP Classifier
y_pred_mlp_classifier = mlp_classifier.predict_proba(X_val)[:, 1]
fpr_mlp_classifier, tpr_mlp_classifier, _ = roc_curve(y_val, y_pred_mlp_classifier)
auc_mlp_classifier = roc_auc_score(y_val, y_pred_mlp_classifier)
ax.plot(fpr_mlp_classifier, tpr_mlp_classifier, lw=2, label=f'MLP Classifier (AUC = {auc_mlp_classifier:.2f})')

ax.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
ax.set_xlim([0.0, 1.0])
ax.set_ylim([0.0, 1.05])
ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')
ax.set_title('ROC Curve - Validation Data for All Classifiers')
ax.legend(loc='lower right')

plt.show()
```
